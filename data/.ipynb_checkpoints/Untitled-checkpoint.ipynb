```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set style for better plots
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 11

# UNIVERSAL DATA LOADING - works on ANY computer
# Assumes CSV is in 'data/' folder next to notebook
data_path = os.path.join('data', 'AB_NYC_2019.csv')

# Try multiple common locations
possible_paths = [
    data_path,  # notebooks/data/AB_NYC_2019.csv
    '../data/AB_NYC_2019.csv',  # data/AB_NYC_2019.csv (if notebook in notebooks/)
    'AB_NYC_2019.csv'  # same folder
]

df = None
for path in possible_paths:
    if os.path.exists(path):
        df = pd.read_csv(path)
        print(f" Loaded from: {path}")
        break

if df is None:
    raise FileNotFoundError(" Dataset not found! Put AB_NYC_2019.csv in 'data/' folder next to notebook.")

print("Dataset shape:", df.shape)
print("\nFirst 3 rows:")
df.head(3)
```

```python
print("Dataset info:")
print(df.info())
print("\nMissing values:")
print(df.isnull().sum())
print("\nBasic statistics:")
print(df.describe())
```

```python
# Remove price = 0 (invalid listings)
df_clean = df[df['price'] > 0].copy()

# Remove extreme price outliers (top 1%)
price_threshold = df_clean['price'].quantile(0.99)
df_clean = df_clean[df_clean['price'] <= price_threshold]

# Convert last_review to datetime
df_clean['last_review'] = pd.to_datetime(df_clean['last_review'], errors='coerce')

# Create calculated columns for analysis
df_clean['price_per_review'] = df_clean['price'] / (df_clean['number_of_reviews'] + 1)  # avoid division by 0

print("After cleaning - shape:", df_clean.shape)
print("\nPrice stats after cleaning:")
print(df_clean['price'].describe())
```

Q1: What is the distribution of Airbnb prices in NYC?

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df_clean['price'], bins=50, kde=True)
plt.title("Distribution of Airbnb Prices in NYC")
plt.xlabel("Price ($)")
plt.ylabel("Number of listings")

plt.subplot(1, 2, 2)
sns.boxplot(y=df_clean['price'])
plt.title("Boxplot of Prices")
plt.ylabel("Price ($)")

plt.tight_layout()
plt.show()
```

Q2: How do average prices differ by room type?

```python
plt.figure(figsize=(8, 5))
price_by_room = df_clean.groupby('room_type')['price'].agg(['mean', 'median', 'count']).round(0)
sns.barplot(data=df_clean, x='room_type', y='price', estimator=np.mean)
plt.title("Average Price by Room Type")
plt.ylabel("Average Price ($)")
plt.xlabel("Room Type")

# Add count labels
for i, row in price_by_room.iterrows():
    plt.text(i, row['mean'] + 10, f'n={int(row["count"])}', ha='center')

plt.show()
print(price_by_room)
```

Q3: Which borough has the highest median prices?

```python
plt.figure(figsize=(10, 6))
price_by_borough = df_clean.groupby('neighbourhood_group')['price'].median().sort_values(ascending=False)
sns.barplot(x=price_by_borough.index, y=price_by_borough.values)
plt.title("Median Price by Borough (Neighbourhood Group)")
plt.ylabel("Median Price ($)")
plt.xlabel("Borough")
plt.xticks(rotation=45)
plt.show()

print("Median prices by borough:\n", price_by_borough.round(0))
```

Q4: Which borough has the highest median prices?

```python
price_by_borough = df_clean.groupby('neighbourhood_group')['price'].median().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=price_by_borough.index, y=price_by_borough.values, 
            hue=price_by_borough.index, palette='viridis', legend=False)
plt.title("Median Price by Borough", fontsize=14, fontweight='bold')
plt.ylabel("Median Price ($)")
plt.xlabel("Borough")
plt.xticks(rotation=45)
for i, v in enumerate(price_by_borough.values):
    plt.text(i, v + 5, f"${v:.0f}", ha='center', fontweight='bold')
plt.tight_layout()
plt.show()

print("Median prices by borough:\n", price_by_borough.round(0))
```

Insight: Manhattan dominates with median $155/night due to prime location. Brooklyn offers value at $90 median. Staten Island has few listings

Q5: Is there a relationship between price and number of reviews?

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(data=df_clean, x='number_of_reviews', y='price', alpha=0.6)
plt.title("Price vs Number of Reviews")
plt.xlabel("Number of Reviews")
plt.ylabel("Price ($)")

plt.subplot(1, 2, 2)
sns.boxplot(data=df_clean[df_clean['number_of_reviews'] <= 100], y='price')
plt.title("Price Distribution (Reviews  100)")
plt.ylabel("Price ($)")

plt.tight_layout()
plt.show()
```

**Insight**: No strong correlation. High-review listings exist across all price ranges, suggesting popularity  luxury.

Q6: How does yearly availability differ by borough?

```python
avail_by_borough = df_clean.groupby('neighbourhood_group')['availability_365'].mean().sort_values()
plt.figure(figsize=(10, 6))
sns.barplot(x=avail_by_borough.index, y=avail_by_borough.values, 
            hue=avail_by_borough.index, palette='coolwarm', legend=False)
plt.title("Average Availability (days/year) by Borough", fontsize=14)
plt.ylabel("Average Availability (days)")
plt.xlabel("Borough")
plt.xticks(rotation=45)
for i, v in enumerate(avail_by_borough.values):
    plt.text(i, v + 5, f"{v:.0f} days", ha='center')
plt.tight_layout()
plt.show()
print("Average availability:\n", avail_by_borough.round(1))
```

Insight: Queens has highest availability (~140 days/year), indicating more flexible hosts. Manhattan listings book up fastest.

Q7: Which hosts have the most listings?

```python
top_hosts = df_clean['host_name'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_hosts.values, y=top_hosts.index, 
            hue=top_hosts.index, palette='plasma', legend=False)
plt.title("Top 10 Hosts by Number of Listings", fontsize=14)
plt.xlabel("Number of Listings")
plt.ylabel("Host Name")
plt.tight_layout()
plt.show()
print("Top hosts:\n", top_hosts)
```

Insight: Jelisa (276 listings) and Luz (278) are professional hosts dominating the market.

Q8: Does requiring more minimum nights correlate with higher price?

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(data=df_clean[df_clean['minimum_nights'] <= 30], 
                x='minimum_nights', y='price', alpha=0.6)
plt.title("Price vs Minimum Nights (30 nights)")
plt.xlabel("Minimum Nights")
plt.ylabel("Price ($)")

plt.subplot(1, 2, 2)
min_nights_price = df_clean.groupby('minimum_nights')['price'].median().head(20)
plt.plot(min_nights_price.index, min_nights_price.values, marker='o', linewidth=2, markersize=6)
plt.title("Median Price by Minimum Nights (top 20)")
plt.xlabel("Minimum Nights")
plt.ylabel("Median Price ($)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Insight: Short stays (1-3 nights) have wide price range. Longer minimums (>7 nights) tend toward higher prices. 

Q9: How do room types distribute across boroughs?

```python
plt.figure(figsize=(12, 6))
ct = pd.crosstab(df_clean['neighbourhood_group'], df_clean['room_type'], normalize='index') * 100
ct.plot(kind='bar', stacked=True)
plt.title("Room Type Distribution by Borough (%)", fontsize=14)
plt.ylabel("Percentage of Listings")
plt.xlabel("Borough")
plt.legend(title="Room Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print(ct.round(1))
```

Insight: Entire homes dominate Brooklyn/Manhattan (60-70%). Private rooms popular everywhere. Shared rooms rare outside Brooklyn.
